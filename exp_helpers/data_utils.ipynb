```python
import pandas as pd
pd.options.display.max_colwidth=500
```

```python
root_dir = '../'
base_dir = root_dir + 'data/magpie/'
data_file = base_dir + 'MAGPIE_filtered_split_typebased.jsonl'
output_file = base_dir + 'processed_MAGPIE_filtered_split_typebased.csv'
```

```python
df_zero_shot = pd.read_json(data_file, lines=True)
df_zero_shot
```

```python
column_subset = ['confidence', 'context', 'idiom', 'label', 'split', 'variant_type']
df_zero_shot = df_zero_shot[column_subset]
df_zero_shot
```

```python
# All examples have exactly 5 sentences, so 3rd sentence is going to contain the Potentially Idiomatic Expression(PIE).
df_zero_shot['context'].map(lambda c: len(c)).value_counts()
```

```python
# Take only the sentence with PIE
df_zero_shot['sentence_0'] = df_zero_shot['context'].map(lambda ctx: ctx[2])
df_zero_shot[['sentence_0', 'idiom']]
```

```python
output_cols = ['sentence_0', 'idiom', 'confidence', 'label', 'split', 'variant_type']
df_zero_shot = df_zero_shot[output_cols]
df_zero_shot.to_csv(output_file, index=False)
print(f'Saved the file to {output_file}')
```
