```python
import pandas as pd
pd.options.display.max_colwidth=500
```

```python
root_dir = '../'
base_dir = root_dir + 'data/magpie/'
data_file = base_dir + 'MAGPIE_filtered_split_typebased.jsonl'
no_context_output_file = base_dir + 'processed_MAGPIE_filtered_split_typebased.csv'
context_output_file = base_dir + 'processed_with_context_MAGPIE_filtered_split_typebased.csv'
```

```python
df_zero_shot = pd.read_json(data_file, lines=True)
df_zero_shot
```

```python
column_subset = ['confidence', 'context', 'idiom', 'label', 'split', 'variant_type', 'offsets']
df_zero_shot = df_zero_shot[column_subset]
df_zero_shot
```

```python
# All examples have exactly 5 sentences, so 3rd sentence is going to contain the Potentially Idiomatic Expression(PIE).
df_zero_shot['context'].map(lambda c: len(c)).value_counts()
```

```python
# Take only the sentence with PIE
df_zero_shot['sentence_0'] = df_zero_shot['context'].map(lambda ctx: ctx[2].replace("\n", " "))
df_zero_shot[['sentence_0', 'idiom']]
```

## Save only the Sentence without context

```python
output_cols = ['sentence_0', 'idiom', 'confidence', 'label', 'split', 'variant_type', 'offsets']
tmp = df_zero_shot[output_cols]
tmp.to_csv(no_context_output_file, index=False)
print(f'Saved the file to {no_context_output_file}')
```

## Save 'context' sentences on one document

```python
# Merge the context sentences into single document with PIE
df_zero_shot['sentence'] = df_zero_shot['context'].map(lambda ctx: " ".join(ctx).replace("\n", " "))
df_zero_shot[['sentence', 'idiom']]
```

```python
output_cols = ['sentence', 'idiom', 'confidence', 'label', 'split', 'variant_type', 'offsets']
tmp = df_zero_shot[output_cols]
tmp.to_csv(context_output_file, index=False)
print(f'Saved the file to {context_output_file}')
```
